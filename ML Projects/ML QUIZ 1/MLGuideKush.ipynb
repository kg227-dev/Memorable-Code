## Removing Variables using Pandas Profiling 

```python
# Delete columns from the dataframe
vals.remove("name")
vals.remove("primary_percent")
vals.remove("general_percent")
vals.remove("party")
vals.remove("presidential_vote")

dataset=dataset.dropna()
```

## Histogram Visualization 

```python
#Plot creation
plt.figure(figsize=(8,5))
plt.tight_layout()

seabornInstance.distplot(dataset['vote'])
mean=dataset['vote'].mean()
median=dataset['vote'].median()

plt.axvline(mean, color='r', linestyle='--')
plt.axvline(median, color='g', linestyle='-')

plt.legend({'Mean':mean,'Median':median})

mmm = pd.DataFrame({'vote': ['mean', 'median'], 'Value': [mean, median]})
display(mmm)
```

## Splitting & Normalization 

```python
#splits data into train and test
#predictors
X = dataset[vals].values

#target variable
y = dataset['vote'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(X_train)
# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Initial Model: ElasticNet Regularization (All Variables) 

```python
ElasticModel = ElasticNetCV(l1_ratio=0.5, eps=0.001, n_alphas=1000, alphas=None, fit_intercept=True, normalize=False, precompute='auto', cv=3, copy_X=True, verbose=0, n_jobs=None)
ElasticVisualizer = AlphaSelection(ElasticModel)
ElasticVisualizer.fit(X_train, y_train)
ElasticVisualizer.show()
```

```python
#Prints ElasticNet Actual vs Predicted in Tabular Format 

print("ElasticNet Predict with transformations:")
clf = linear_model.ElasticNet(alpha=.003, fit_intercept=True, normalize=False, precompute=False, copy_X=True, max_iter=1000, tol=0.0001,  selection='cyclic')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
pd.set_option('display.max_rows', 100)
coefs = pd.DataFrame({ 'Variables': vals, 'Coefficients': clf.coef_})
display(coefs)

print("Intercept: " + str(clf.intercept_))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

## Feature Selection: Random Forest

```python
yvar = "vote"
```

```python
#Converts to Ints
lab_enc = preprocessing.LabelEncoder()
rfData = pd.DataFrame()
for i in dataset.columns:
    encoded = lab_enc.fit_transform(dataset[i])
    rfData[i] = encoded
```

```python
X = rfData.loc[:, vals]
Y = rfData[yvar]

RF_train_x, RF_test_x, RF_train_y, RF_test_y = train_test_split(X, Y, test_size=0.2, random_state=0)

rf = RandomForestClassifier(n_estimators=100, oob_score=True)
rf = rf.fit(RF_train_x, RF_train_y)
RF_pred_y=rf.predict(RF_test_x)

print("Random Forest Accuracy:",metrics.accuracy_score(RF_test_y, RF_pred_y))
```

## Variable Importances: Random Forest 

```python
#%%
# what features were important?
importances = rf.feature_importances_
std = np.std([tree.feature_importances_ for tree in rf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

important_features = pd.Series(data=importances,index=X.columns)
important_features.sort_values(ascending=False,inplace=True)
i = 1

print("Feature                       Percentage Scale")

for feature in important_features:
    print("{:2}. {:30} {}%".format(i, important_features.index[i-1], np.round(feature*100/important_features[0], 2)))
    i+=1
```

```python
#describes values chosen 
#vals = "t0,t1,t2,t3,t4,t5,t6,t7,t8,t9,t10,t11,t12,t13,t14,t15,t16,t17,t18,t19,t20,t21,t22,t23,t24,t25,t26,t27,t28,t29,t30,t31,t32,t33,t34,t35,t36,t37,t38,t39,t40,t41,t42,t43,t44,t45,t46,t47,t48,t49"
values = "dwnom1,ideo5,dwnom2,general_margin,presidential_margin,un,primary_margin,tenure_at_beginning_of_term,t41,t17,t30,t24,t8,incumbent"
values = values.split(",") 

X = dataset[values].values

#target variable
y = dataset['vote'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

scaler = StandardScaler()
# Fit on training set only.
scaler.fit(X_train)
# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Prints ElasticNet Actual vs Predicted in Tabular Format 

print("ElasticNet Predict with transformations:")
clf = linear_model.ElasticNet(alpha=0.001, fit_intercept=True, normalize=False, precompute=False, copy_X=True, max_iter=1000, tol=0.0001,  selection='cyclic')
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

coefs = pd.DataFrame({ 'Variables': values, 'Coefficients': clf.coef_})
display(coefs)

print("Intercept: " + str(clf.intercept_))
```
